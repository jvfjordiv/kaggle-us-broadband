# 1. Data mining

First of all, we read it:

```python
#loading
pd.set_option('display.max_columns', None)
csv = pd.read_csv('../../data/broadband_access.csv')
print("We have", csv.shape[0], "entries.")
print("We have", csv.shape[1], "attributes.")
print(csv.head())
```

We have info for every county in the United States, which gives us 3142 tuples. Each have their own stadistics about population, ratios for unemployment, health insurance, broadband connection at home, broadband availability, number of providers...

However, some attributes names sound unnecessary and redundant, like "lowest broadband cost per month (from BBN set)" (`price_bbn`) and "lowest cost per month (from ILMS set)" (`broad_cost`). Looks like we have duplicate data which comes from two sources. We also appear to have NaN and empty values. But before we treat them, let's check this correlation matrix. Maybe those attributes can actually give us some useful data.

```python
correlacio = csv.corr()
plt.figure(figsize=(13,13))
ax = sns.heatmap(correlacio, annot=True, linewidths=.5)
```

At first sight, we see, for example, that `population` and `population_bbn` are duplicates, and `no_internet` contains the inverse data of `home_broad`. So we can remove one of them.

Other values that have strong correlation include `no_comp` (percent with no home computer) with `no_internet` (percent with no home internet), `poverty` (poverty rate) with `SNAP` (percent of people who received Supplemental Nutrition Assistance Program), `wired_bbn` (number of Wired providers) with `provide_bbn` (number of providers), and `all25_bbn` (providers offering high speed connections) with `provide_bbn`.

Apparently, the `no_internet` (percent with no home internet) attribute is the one that interests us the most.
